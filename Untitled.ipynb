```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
confirmed_df = pd.read_csv('confirmed.csv')
deaths_df = pd.read_csv('deaths.csv')
recovered_df = pd.read_csv('recovered.csv')

print("DataFrames loaded successfully.")
```

```python
def reshape_dataframe(df, case_type):
    """
    Reshapes the wide format dataframe into a long format.
    """
    # Exclude the `...` column as it is not a valid date
    date_columns = [col for col in df.columns if '/' in col]
    id_vars = ['Province/State', 'Country/Region', 'Lat', 'Long']
    
    # Melt the dataframe
    df_long = df.melt(id_vars=id_vars, value_vars=date_columns, var_name='Date', value_name=case_type)
    
    # Convert 'Date' to datetime objects
    df_long['Date'] = pd.to_datetime(df_long['Date'])
    
    # Aggregate data by Country/Region and Date
    df_long = df_long.groupby(['Country/Region', 'Date']).sum().reset_index()
    return df_long

# Apply the function to each dataframe
confirmed_long = reshape_dataframe(confirmed_df, 'Confirmed')
deaths_long = reshape_dataframe(deaths_df, 'Deaths')
recovered_long = reshape_dataframe(recovered_df, 'Recovered')

print("DataFrames reshaped and cleaned successfully.")
print("\nSample of reshaped data:")
print(confirmed_long.head())
```

```python
# Merge the dataframes
full_data = pd.merge(confirmed_long, deaths_long, on=['Country/Region', 'Date'], how='outer')
full_data = pd.merge(full_data, recovered_long, on=['Country/Region', 'Date'], how='outer')

# Fill missing recovered values with 0
full_data['Recovered'] = full_data['Recovered'].fillna(0)
full_data = full_data.fillna(0)

# Sort the data by country and date
full_data.sort_values(['Country/Region', 'Date'], inplace=True)

# Calculate active cases
full_data['Active'] = full_data['Confirmed'] - full_data['Deaths'] - full_data['Recovered']

# Calculate daily new cases, deaths, and recoveries
full_data['Daily_Confirmed'] = full_data.groupby('Country/Region')['Confirmed'].diff().fillna(full_data['Confirmed'])
full_data['Daily_Deaths'] = full_data.groupby('Country/Region')['Deaths'].diff().fillna(full_data['Deaths'])
full_data['Daily_Recovered'] = full_data.groupby('Country/Region')['Recovered'].diff().fillna(full_data['Recovered'])

# Calculate mortality and recovery rates
full_data['Mortality_Rate'] = (full_data['Deaths'] / full_data['Confirmed']) * 100
full_data['Recovery_Rate'] = (full_data['Recovered'] / full_data['Confirmed']) * 100

print("Data merged and new metrics calculated.")
print("\nSample of final data:")
print(full_data.head())
```

```python

# Select a few countries for comparison
countries_of_interest = ['Afghanistan', 'Australia', 'Albania', 'Algeria', 'Austria']
data_for_plot = full_data[full_data['Country/Region'].isin(countries_of_interest)]

# Line plots for cumulative cases
plt.figure(figsize=(15, 8))
sns.lineplot(data=data_for_plot, x='Date', y='Confirmed', hue='Country/Region')
plt.title('Cumulative Confirmed Cases Over Time')
plt.ylabel('Cumulative Confirmed Cases')
plt.xlabel('Date')
plt.grid(True)
plt.show()

# Area chart for a single country (e.g., Afghanistan)
single_country = 'Afghanistan'
single_country_data = full_data[full_data['Country/Region'] == single_country]
single_country_data_stacked = single_country_data.set_index('Date')[['Confirmed', 'Deaths', 'Recovered']]

plt.figure(figsize=(15, 8))
single_country_data_stacked.plot(kind='area', stacked=True, figsize=(15, 8))
plt.title(f'Stacked Area Chart of COVID-19 Cases in {single_country}')
plt.ylabel('Number of Cases')
plt.xlabel('Date')
plt.grid(True)
plt.legend(title='Case Type')
plt.show()

# Heatmap for daily new cases
heatmap_data = full_data.pivot_table(index='Country/Region', columns='Date', values='Daily_Confirmed', fill_value=0)
heatmap_data = heatmap_data.loc[countries_of_interest]

plt.figure(figsize=(20, 10))
sns.heatmap(heatmap_data, cmap='YlGnBu', linewidths=0.5, linecolor='black')
plt.title('Heatmap of Daily New Confirmed Cases')
plt.xlabel('Date')
plt.ylabel('Country/Region')
plt.tight_layout()
plt.show()
```
